```python
%sql
USE CATALOG workspace;
USE SCHEMA default;
```

```python
%sql
DROP TABLE IF EXISTS workspace.default.courses;
```

```python
%sql
-- Schritt 1: Leere Tabelle im Unity Catalog anlegen
CREATE TABLE workspace.default.courses (
  course_id BIGINT, 
  title STRING,
  instructor STRING,
  category STRING,
  price BIGINT 
) USING DELTA;
```

```python
%sql
-- Schritt 2: Daten laden
COPY INTO workspace.default.courses
FROM '/Volumes/workspace/default/volume/courses.json'
FILEFORMAT = JSON
FORMAT_OPTIONS ('inferSchema' = 'true') -- Spark versucht die Datentypen aus dem JSON zu lesen
COPY_OPTIONS ('mergeSchema' = 'true'); -- Erlaubt neue Spalten, falls das JSON wächst
```
